```python
# getting all the imdb movie ids
# imdb uses single page pagination so i need to do it manually
# (by manually i mean mechanicalsoup should open the page and click the button manually)

# i think i need to use selenium

from math import ceil
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

import time 
  
from selenium.webdriver.common.keys import Keys 

browser = webdriver.Chrome()
browser.maximize_window()

browser.get("https://www.imdb.com/search/title/?title_type=feature")

time.sleep(5)

#print(browser.find_elements(By.CLASS_NAME, "ipc-title-link-wrapper"))

browser.implicitly_wait(5)
```

```python
actions = ActionChains(browser)
asd = browser.find_element(By.CLASS_NAME, "ipc-page-grid").find_elements(By.CLASS_NAME, "ipc-page-grid__item")[1]
number_of_movies_str = asd.text.split("\n")[0].split(" ")[2].replace(",", "")
print(number_of_movies_str)
number_of_movies = int(number_of_movies_str)
print(number_of_movies)
number_of_max_clicks = ceil(number_of_movies / 50)

print("there are " + number_of_movies_str + " movies. we need to click " + str(number_of_max_clicks) + " in total. haha")


#b = browser.find_elements(By.CLASS_NAME, "ipc-see-more__button")
b = browser.find_elements(By.CLASS_NAME, "ipc-see-more")
print(b)
#actions.move_to_element_with_offset(b[0], 0, 10).click().perform()
#actions.move_to_element(b[0]).move_by_offset(100, 100).perform()
#browser.execute_script("arguments[0].scrollIntoView();", b[0])
for i in range(number_of_max_clicks):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(2)
    b[0].click()
    time.sleep(2)
    print("[LOG] " + str(i) + " out of " + str(number_of_max_clicks) + " completed. %" + str(i / number_of_max_clicks))
```
